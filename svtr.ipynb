## Import Stuff

```python
import pandas as pd
import numpy as np
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.metrics import r2_score
from sklearn.feature_selection import SelectPercentile, SelectKBest, f_regression, SelectFromModel, chi2
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.ensemble import IsolationForest
```

```python
def eval(estimator, domain, target):
    return r2_score(y_pred=estimator.predict(domain), y_true=target.ravel())
```

# Read Datasets

```python
df = pd.read_csv('X_train.csv')
yf = pd.read_csv('y_train.csv')
tf = pd.read_csv('X_test.csv')
tarray = tf.values
array = df.values
arrayy = yf.values
T = tarray[:,1:]
X = array[:,1:]
y = arrayy[:,1:]
it = IterativeImputer(n_nearest_features=15, random_state=0)
X = it.fit_transform(X)
T = it.fit_transform(T)
```

# Split into Test and Training Sets

```python
Xt, Xv, yt, yv = train_test_split(X, y, test_size=0.2, random_state=0)
```

# Create Pipeline



```python
pipe = Pipeline([('imputer', IterativeImputer(n_nearest_features=15, random_state=0)),
                 ('scaler', StandardScaler()), ('feature selector', SelectPercentile(f_regression)), 
                 ('svr', SVR(kernel='rbf', gamma='auto'))])
```

# Grid search everything

```python
params = {'feature selector__percentile': [10, 15, 20, 30], 'svr__C': range(20, 80, 5), 'svr__epsilon': [0.001, 0.01, 0.1, 0.5, 1, 1.5, 10]}
gsrc = GridSearchCV(pipe, params)
gsrc.fit(Xt, yt.ravel())
print(gsrc.cv_results_)
print(gsrc.best_score_)
bp = gsrc.best_params_
print(bp)
```

```python
eval(gsrc, Xv, yv)
```

```python
eval(gsrc2, Xv2, yv)
```
